<b><font size = 3>1. Calculating the win/loss ratio's correlation with the population of the city it is in for the NHL using 2018 data</font></b>

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import re

nhl_df=pd.read_csv("assets/nhl.csv")
cities=pd.read_html("assets/wikipedia_data.html")[1]
cities=cities.iloc[:-1,[0,3,5,6,7,8]]

def nhl_correlation(): 
    nhl_df=pd.read_csv("assets/nhl.csv")
    nhl_df=nhl_df[nhl_df['year']==2018]
    nhl_df=nhl_df[{'W', 'L', 'team'}]
    nhl_df=nhl_df.drop([0,9,18,26])
    nhl_df.reset_index()
    nhl_df["W"] = nhl_df["W"].astype(str).astype(int)
    nhl_df["L"] = nhl_df["L"].astype(str).astype(int)
    nhl_df['Ratio']=nhl_df['W']/(nhl_df['W']+nhl_df['L'])
    nhl_df
    cities=pd.read_html("assets/wikipedia_data.html")[1]
    cities=cities.iloc[:-1,[0,3,5,6,7,8]]
    citiesn=cities[{'Metropolitan area', 'NHL'}]
    citiesn['NHL'].iloc[0]="Rangers Islanders Devils"
    citiesn['NHL'].iloc[1]="Kings Ducks"
    citiesn['NHL'].iloc[2]="Sharks"
    citiesn['NHL'].iloc[6]="Flyers"
    citiesn['NHL'].iloc[8]="Wild"
    citiesn['NHL'].iloc[9]="Avalanche"
    citiesn['NHL'].iloc[17]="Penguins"
    citiesn['NHL'].iloc[22]="Blues"
    citiesn['NHL'].iloc[39]="Jets"
    for team in nhl_df['team']:
        nhl_df['team'][nhl_df['team']==team]=team.replace("*", "")
    nhl_df["City"]=1234
    lt=[]
    for t in citiesn['NHL']:
        for ch in t.split(" "):
            for team in nhl_df["team"]:
                if team.find(ch)!=-1:
                    city=citiesn["Metropolitan area"][citiesn['NHL']==t].to_numpy()
                    ratio=nhl_df['Ratio'][nhl_df['team']==team].to_numpy()
                    d={'Ratio': ratio[0], 'Team': team, 'City': city[0]}
                    lt.append(d)
    new_lt = []
    for i in lt:
        if i not in new_lt:
            new_lt.append(i)
    df=pd.DataFrame(new_lt)
    df
    groupby_column = 'City'
    aggregate_column = 'Ratio'
    agg_df = df.groupby(groupby_column).mean()
    citiesp=cities[{'Metropolitan area', 'Population (2016 est.)[8]'}]
    citiesp.rename(columns={'Metropolitan area': 'City'}, inplace=True)
    citiesp=citiesp.set_index('City')
    dfp=pd.merge(agg_df, citiesp, how='left', left_index=True, right_index=True)
    dfp["Population (2016 est.)[8]"] = dfp["Population (2016 est.)[8]"].astype(str).astype(int)
    
    population_by_region = dfp['Population (2016 est.)[8]'].to_numpy()
    win_loss_by_region = dfp['Ratio'].to_numpy() 
    
    return (stats.pearsonr(population_by_region, win_loss_by_region))[0]
print(nhl_correlation())
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import re

nfl_df=pd.read_csv("assets/nfl.csv")
cities=pd.read_html("assets/wikipedia_data.html")[1]
cities=cities.iloc[:-1,[0,3,5,6,7,8]]

def nfl_correlation(): 
    # YOUR CODE HERE
    nhl_df=pd.read_csv("assets/nfl.csv")
    nhl_df=nhl_df[nhl_df['year']==2018]
    nhl_df=nhl_df[{'W', 'L', 'team'}]
    nhl_df=nhl_df.drop([0,5,10,15,20,25, 30 , 35], axis=0)
    nhl_df.reset_index()
    nhl_df["W"] = nhl_df["W"].astype(str).astype(int)
    nhl_df["L"] = nhl_df["L"].astype(str).astype(int)
    nhl_df['Ratio']=nhl_df['W']/(nhl_df['W']+nhl_df['L'])
    nhl_df
    cities=pd.read_html("assets/wikipedia_data.html")[1]
    cities=cities.iloc[:-1,[0,3,5,6,7,8]]
    citiesn=cities[{'Metropolitan area', 'NFL', 'Population (2016 est.)[8]'}]
    citiesn['NFL'].iloc[0]="Giants Jets"
    citiesn['NFL'].iloc[1]="Rams Chargers"
    citiesn['NFL'].iloc[2]="49ers Raiders"
    citiesn['NFL'].iloc[3]="Bears"
    citiesn['NFL'].iloc[7]="Patriots"
    citiesn['NFL'].iloc[14]="Texans"
    citiesn['NFL'].iloc[18]="Browns"
    citiesn['NFL'].iloc[23]="Ravens"
    citiesn['NFL'].iloc[29]="Bills"
    for team in nhl_df['team']:
        nhl_df['team'][nhl_df['team']==team]=team.replace("*", "")
    nhl_df["City"]=1234
    lt=[]
    for t in citiesn['NFL']:
        for ch in t.split(" "):
            for team in nhl_df["team"]:
                if team.find(ch)!=-1:
                    city=citiesn["Metropolitan area"][citiesn['NFL']==t].to_numpy()
                    pop=citiesn["Population (2016 est.)[8]"][citiesn['NFL']==t].to_numpy()
                    ratio=nhl_df['Ratio'][nhl_df['team']==team].to_numpy()
                    win=nhl_df['W'][nhl_df['team']==team].to_numpy()
                    loss=nhl_df['L'][nhl_df['team']==team].to_numpy()
                    d={'W': win[0], 'L': loss[0], 'Team': team, 'City': city[0], 'Population': pop[0]}
                    lt.append(d)
    new_lt = []
    for i in lt:
        if i not in new_lt:
            new_lt.append(i)
    df=pd.DataFrame(new_lt)
    groupby_column = 'City'
    aggregate_column = 'Ratio'
    agg_df = df.groupby(groupby_column).sum()
    citiesp=cities[{'Metropolitan area', 'Population (2016 est.)[8]'}]
    citiesp.rename(columns={'Metropolitan area': 'City'}, inplace=True)
    citiesp=citiesp.set_index('City')
    dfp=pd.merge(agg_df, citiesp, how='left', left_index=True, right_index=True)
    dfp['Ratio']=dfp['W']/(dfp['W']+dfp['L'])
    dfp["Population (2016 est.)[8]"] = dfp["Population (2016 est.)[8]"].astype(str).astype(int)
    df["Population"] = df["Population"].astype(str).astype(int)
    df['Ratio']=df['W']/(df['W']+df['L'])
    
    population_by_region = dfp["Population (2016 est.)[8]"] 
    win_loss_by_region = dfp["Ratio"] 

    return stats.pearsonr(population_by_region, win_loss_by_region)[0]
print(nfl_correlation())
```
